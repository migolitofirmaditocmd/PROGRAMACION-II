## Parte 1: EstadÃ­sticas BÃ¡sicas con Python Puro

### Medidas de Tendencia Central

```python
# Datos de ejemplo: ventas semanales
ventas = [1200, 1500, 1100, 1800, 1600, 1400, 1700]

# 1. Media (promedio)
media = sum(ventas) / len(ventas)
print(f"Media: ${media:.2f}")

# 2. Mediana (valor central)
ventas_ordenadas = sorted(ventas)
n = len(ventas_ordenadas)
if n % 2 == 0:
    # Si hay cantidad par, promedio de los dos valores centrales
    mediana = (ventas_ordenadas[n//2 - 1] + ventas_ordenadas[n//2]) / 2
else:
    # Si es impar, el valor central
    mediana = ventas_ordenadas[n//2]
print(f"Mediana: ${mediana:.2f}")

# 3. Moda (valor mÃ¡s frecuente)
from collections import Counter
contador = Counter(ventas)
moda = contador.most_common(1)[0][0]
print(f"Moda: ${moda:.2f}")
```

### Medidas de DispersiÃ³n

```python
# Usar los mismos datos de ventas
print(f"Datos: {ventas}\n")

# 1. Rango (mÃ¡ximo - mÃ­nimo)
rango = max(ventas) - min(ventas)
print(f"Valor mÃ­nimo: ${min(ventas)}")
print(f"Valor mÃ¡ximo: ${max(ventas)}")
print(f"Rango: ${rango}")

# 2. Varianza
media = sum(ventas) / len(ventas)
varianza = sum((x - media) ** 2 for x in ventas) / len(ventas)
print(f"\nVarianza: {varianza:.2f}")

# 3. DesviaciÃ³n estÃ¡ndar
desviacion_estandar = varianza ** 0.5
print(f"DesviaciÃ³n estÃ¡ndar: ${desviacion_estandar:.2f}")

# 4. Coeficiente de variaciÃ³n (dispersiÃ³n relativa)
coef_variacion = (desviacion_estandar / media) * 100
print(f"Coeficiente de variaciÃ³n: {coef_variacion:.2f}%")
```

### Crear Funciones Reutilizables

```python
def calcular_media(datos):
    """Calcula la media de una lista de nÃºmeros."""
    return sum(datos) / len(datos)


def calcular_mediana(datos):
    """Calcula la mediana de una lista de nÃºmeros."""
    ordenados = sorted(datos)
    n = len(ordenados)
    if n % 2 == 0:
        return (ordenados[n//2 - 1] + ordenados[n//2]) / 2
    else:
        return ordenados[n//2]


def calcular_varianza(datos):
    """Calcula la varianza de una lista de nÃºmeros."""
    media = calcular_media(datos)
    return sum((x - media) ** 2 for x in datos) / len(datos)


def calcular_desviacion_estandar(datos):
    """Calcula la desviaciÃ³n estÃ¡ndar de una lista de nÃºmeros."""
    return calcular_varianza(datos) ** 0.5


def resumen_estadistico(datos, nombre="Datos"):
    """Genera un resumen estadÃ­stico completo.
    
    Args:
        datos (list): Lista de valores numÃ©ricos
        nombre (str): Nombre descriptivo de los datos
    
    Returns:
        dict: Diccionario con todas las estadÃ­sticas
    """
    resumen = {
        "n": len(datos),
        "min": min(datos),
        "max": max(datos),
        "rango": max(datos) - min(datos),
        "media": calcular_media(datos),
        "mediana": calcular_mediana(datos),
        "varianza": calcular_varianza(datos),
        "desv_est": calcular_desviacion_estandar(datos)
    }
    
    return resumen


def imprimir_resumen(resumen, nombre="Datos"):
    """Imprime un resumen estadÃ­stico de forma legible."""
    print("=" * 50)
    print(f"RESUMEN ESTADÃSTICO: {nombre}")
    print("=" * 50)
    print(f"NÃºmero de observaciones: {resumen['n']}")
    print(f"MÃ­nimo:                  {resumen['min']:.2f}")
    print(f"MÃ¡ximo:                  {resumen['max']:.2f}")
    print(f"Rango:                   {resumen['rango']:.2f}")
    print(f"Media:                   {resumen['media']:.2f}")
    print(f"Mediana:                 {resumen['mediana']:.2f}")
    print(f"Varianza:                {resumen['varianza']:.2f}")
    print(f"DesviaciÃ³n estÃ¡ndar:     {resumen['desv_est']:.2f}")
    print("=" * 50)


# Probar funciones
ventas = [1200, 1500, 1100, 1800, 1600, 1400, 1700]
resumen = resumen_estadistico(ventas, "Ventas Semanales")
imprimir_resumen(resumen, "Ventas Semanales")
```

### Cargar y Analizar Datos del CSV

```python
import csv

def cargar_datos_ventas(archivo):
    """Carga datos de ventas desde CSV.
    
    Args:
        archivo (str): Ruta del archivo CSV
    
    Returns:
        list: Lista de diccionarios con datos de productos
    """
    productos = []
    
    with open(archivo, "r", encoding="utf-8") as f:
        lector = csv.DictReader(f)
        
        for fila in lector:
            producto = {
                "producto": fila["producto"],
                "precio": float(fila["precio"]),
                "cantidad_vendida": int(fila["cantidad_vendida"]),
                "categoria": fila["categoria"],
                "ingreso_total": float(fila["precio"]) * int(fila["cantidad_vendida"])
            }
            productos.append(producto)
    
    return productos


# Cargar datos
productos = cargar_datos_ventas("ventas_productos.csv")

print(f"Se cargaron {len(productos)} productos\n")
print("Primeros 3 productos:")
for p in productos[:3]:
    print(p)
```

## Parte 3: AnÃ¡lisis por CategorÃ­as

```python
def agrupar_por_categoria(productos):
    """Agrupa productos por categorÃ­a.
    
    Args:
        productos (list): Lista de diccionarios de productos
    
    Returns:
        dict: Diccionario donde cada clave es una categorÃ­a
    """
    categorias = {}
    
    for p in productos:
        cat = p["categoria"]
        if cat not in categorias:
            categorias[cat] = []
        categorias[cat].append(p)
    
    return categorias


def analizar_categoria(productos_categoria):
    """Analiza estadÃ­sticas de una categorÃ­a.
    
    Args:
        productos_categoria (list): Lista de productos de una categorÃ­a
    
    Returns:
        dict: EstadÃ­sticas de la categorÃ­a
    """
    ingresos = [p["ingreso_total"] for p in productos_categoria]
    cantidades = [p["cantidad_vendida"] for p in productos_categoria]
    
    return {
        "num_productos": len(productos_categoria),
        "ingreso_total": sum(ingresos),
        "ingreso_promedio": calcular_media(ingresos),
        "cantidad_total": sum(cantidades),
        "cantidad_promedio": calcular_media(cantidades)
    }


# Agrupar productos
categorias = agrupar_por_categoria(productos)

# Analizar cada categorÃ­a
print("=" * 60)
print("ANÃLISIS POR CATEGORÃA")
print("=" * 60)

for nombre_cat, prods in categorias.items():
    stats = analizar_categoria(prods)
    
    print(f"\nðŸ“ {nombre_cat}")
    print("-" * 60)
    print(f"  NÃºmero de productos:       {stats['num_productos']}")
    print(f"  Ingreso total:             ${stats['ingreso_total']:,.2f}")
    print(f"  Ingreso promedio:          ${stats['ingreso_promedio']:,.2f}")
    print(f"  Unidades vendidas (total): {stats['cantidad_total']}")
    print(f"  Unidades vendidas (prom):  {stats['cantidad_promedio']:.1f}")
```

## Parte 4: Top Productos (Ranking)

```python
def top_productos_por_ingreso(productos, n=5):
    """Encuentra los productos con mayor ingreso.
    
    Args:
        productos (list): Lista de productos
        n (int): NÃºmero de productos a retornar
    
    Returns:
        list: Top n productos ordenados por ingreso
    """
    ordenados = sorted(productos, key=lambda p: p["ingreso_total"], reverse=True)
    return ordenados[:n]


def top_productos_por_cantidad(productos, n=5):
    """Encuentra los productos mÃ¡s vendidos por cantidad.
    
    Args:
        productos (list): Lista de productos
        n (int): NÃºmero de productos a retornar
    
    Returns:
        list: Top n productos ordenados por cantidad vendida
    """
    ordenados = sorted(productos, key=lambda p: p["cantidad_vendida"], reverse=True)
    return ordenados[:n]


# Top 5 por ingreso
print("=" * 60)
print("ðŸ† TOP 5 PRODUCTOS POR INGRESO TOTAL")
print("=" * 60)

top_ingreso = top_productos_por_ingreso(productos, 5)
for i, p in enumerate(top_ingreso, 1):
    print(f"{i}. {p['producto']:20} - ${p['ingreso_total']:>10,.2f}")

# Top 5 por cantidad
print("\n" + "=" * 60)
print("ðŸ† TOP 5 PRODUCTOS POR CANTIDAD VENDIDA")
print("=" * 60)

top_cantidad = top_productos_por_cantidad(productos, 5)
for i, p in enumerate(top_cantidad, 1):
    print(f"{i}. {p['producto']:20} - {p['cantidad_vendida']:>4} unidades")
```

## Parte 6: Generar Reporte Completo

```python
def generar_reporte_completo(productos, guardar_archivo=True):
    """Genera un reporte estadÃ­stico completo de ventas.
    
    Args:
        productos (list): Lista de productos
        guardar_archivo (bool): Si True, guarda el reporte en un archivo
    """
    lineas = []
    
    # Encabezado
    lineas.append("=" * 70)
    lineas.append("REPORTE ESTADÃSTICO DE VENTAS")
    lineas.append("Universidad de Guadalajara")
    lineas.append("=" * 70)
    lineas.append("")
    
    # Resumen general
    lineas.append("ðŸ“Š RESUMEN GENERAL")
    lineas.append("-" * 70)
    lineas.append(f"Total de productos analizados: {len(productos)}")
    
    ingreso_total = sum(p["ingreso_total"] for p in productos)
    lineas.append(f"Ingreso total:                 ${ingreso_total:,.2f}")
    
    cantidad_total = sum(p["cantidad_vendida"] for p in productos)
    lineas.append(f"Unidades vendidas (total):     {cantidad_total}")
    lineas.append("")
    
    # EstadÃ­sticas de precios
    precios = [p["precio"] for p in productos]
    lineas.append("ðŸ’° ESTADÃSTICAS DE PRECIOS")
    lineas.append("-" * 70)
    lineas.append(f"Precio mÃ­nimo:      ${min(precios):,.2f}")
    lineas.append(f"Precio mÃ¡ximo:      ${max(precios):,.2f}")
    lineas.append(f"Precio promedio:    ${calcular_media(precios):,.2f}")
    lineas.append(f"Precio mediano:     ${calcular_mediana(precios):,.2f}")
    lineas.append("")
    
    # AnÃ¡lisis por categorÃ­a
    lineas.append("ðŸ“ ANÃLISIS POR CATEGORÃA")
    lineas.append("-" * 70)
    
    categorias = agrupar_por_categoria(productos)
    for nombre_cat, prods in categorias.items():
        stats = analizar_categoria(prods)
        lineas.append(f"\n  {nombre_cat}:")
        lineas.append(f"    Productos:        {stats['num_productos']}")
        lineas.append(f"    Ingreso total:    ${stats['ingreso_total']:,.2f}")
        lineas.append(f"    % del total:      {(stats['ingreso_total']/ingreso_total*100):.1f}%")
    
    lineas.append("")
    
    # Top 3 productos
    lineas.append("ðŸ† TOP 3 PRODUCTOS POR INGRESO")
    lineas.append("-" * 70)
    top3 = top_productos_por_ingreso(productos, 3)
    for i, p in enumerate(top3, 1):
        lineas.append(f"{i}. {p['producto']:20} ${p['ingreso_total']:>10,.2f}")
    
    lineas.append("")
    lineas.append("=" * 70)
    lineas.append("Fin del reporte")
    lineas.append("=" * 70)
    
    # Unir todas las lÃ­neas
    reporte = "\n".join(lineas)
    
    # Imprimir
    print(reporte)
    
    # Guardar en archivo si se solicita
    if guardar_archivo:
        with open("reporte_ventas.txt", "w", encoding="utf-8") as f:
            f.write(reporte)
        print("\nâœ“ Reporte guardado en 'reporte_ventas.txt'")


# Generar reporte
generar_reporte_completo(productos, guardar_archivo=True)
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Crear datos de ejemplo
datos_calif = """nombre,matematicas,fisica,quimica,historia
Ana GarcÃ­a,85,90,88,92
Luis MartÃ­nez,92,87,90,85
MarÃ­a LÃ³pez,88,92,85,90
Carlos PÃ©rez,90,88,92,87
Elena RodrÃ­guez,87,85,89,91
Pedro SÃ¡nchez,91,89,87,88
Laura JimÃ©nez,89,91,90,89
Diego Torres,86,88,86,90
"""

# ESCRIBA SU SOLUCIÃ“N AQUÃ

def actualizar_base_de_datos(BOLETA,ruta,campos):
    with open(ruta,"w",encoding="utf-8",newline="") as f:
        f.seek(0)
        escritor = csv.DictWriter(f,fieldnames=campos)
        escritor.writeheader()
        for alumno in BOLETA:
            escritor.writerow(alumno)


def abrir_generar_archivo(ruta):
    with open(ruta, "a+", encoding="utf-8", newline="") as f:
        f.seek(0)
        contenido = f.read()

        if contenido == "":
            f.write(datos_calif)
            f.seek(0)
        f.seek(0)
        lector = csv.DictReader(f)
        calificaciones = list(lector)

    return calificaciones

def boleta_individual(alumnos):
    materias = ["Matematicas","Fisica","Quimica","Historia","Promedio"]
    calificaciones = []
    lista_de_alumnos = [alumno["nombre"] for alumno in alumnos]
    encontrado = True
    while encontrado:
        busqueda = input(f"Nombre del estudiante Lista de estudiantes:{lista_de_alumnos} (N para salir): ")
        if busqueda.lower() == "n":
            return

        for alumno in alumnos:
            if alumno["nombre"] == busqueda:
                encontrado = True
                calificaciones = [
                    alumno["matematicas"],
                    alumno["fisica"],
                    alumno["quimica"],
                    alumno["historia"],
                    float(alumno["promedio"])
                ]
                print("Nombre:", alumno["nombre"])
                print("Calificaciones:")
                print("MatemÃ¡ticas:", alumno["matematicas"])
                print("FÃ­sica:", alumno["fisica"])
                print("QuÃ­mica:", alumno["quimica"])
                print("Historia:", alumno["historia"])
                print("Promedio:", alumno["promedio"])
                plt.bar(materias, calificaciones, color="skyblue")
                plt.title(f"Boleta de {alumno['nombre']}")
                plt.ylabel("CalificaciÃ³n")
                plt.ylim(0, 100)
                for i, cal in enumerate(calificaciones):
                    plt.text(i, cal + 1, str(cal), ha="center")

                plt.show()
                break
def promedio(alumno):
    promiedin = float((alumno["matematicas"] + alumno["fisica"] + alumno["quimica"] + alumno["historia"])/4)
    return promiedin

def boleta_global(BOLETA):
    tabla_alumnos = []
    for alumno in BOLETA:
        fila = {
            "nombre": alumno["nombre"],
            "matematicas": alumno["matematicas"],
            "fisica": alumno["fisica"],
            "quimica": alumno["quimica"],
            "historia": alumno["historia"],
            "promedio": alumno["promedio"]
        }
        tabla_alumnos.append(fila)
    lista_estad = estadisticas_materias(BOLETA)
    estadisticas_por_materia = lista_estad[0]
    materia_mas_facil = lista_estad[1][0]
    materia_mas_dificil = lista_estad[1][1]

    tabla_global = {
        "boletas_alumnos": tabla_alumnos,
        "estadisticas_materias": estadisticas_por_materia,
        "materia_mas_facil": materia_mas_facil,
        "materia_mas_dificil": materia_mas_dificil
    }

    return tabla_global

def estadisticas_materias(boleta):
    mates = []
    fis = []
    quim = []
    his = []
    materias_globales = [{k:v for (k,v) in alumno.items() if k != "nombre" and k != "promedio"} for alumno in boleta]
    for materias in materias_globales:
        mates.append(materias["matematicas"])
        fis.append(materias["fisica"])
        quim.append(materias["quimica"])
        his.append(materias["historia"])
    facilDificil = facil_dificil(matematicas=mates,fisica=fis,quimica=quim,historia=his)
    return [{"matematicas":{"promedio": sum(mates)/len(mates),"max":max(mates),"min":min(mates)},
            "fisica":{"promedio": sum(fis)/len(fis),"max":max(fis),"min":min(fis)},
            "quimica":{"promedio": sum(quim)/len(quim),"max":max(quim),"min":min(quim)},
            "historia":{"promedio": sum(his)/len(his),"max":max(his),"min":min(his)}},facilDificil]


def facil_dificil(matematicas,fisica,quimica,historia):
    promedios = {"matematicas":sum(matematicas)/len(matematicas),
                 "fisica": sum(fisica)/len(fisica),
                 "quimica": sum(quimica)/len(quimica),
                 "historia":sum(historia)/len(historia)}
    nombre_facil = max(promedios, key=promedios.get)
    materia_mas_facil = {nombre_facil: promedios[nombre_facil]}
    nombre_dificil = min(promedios, key=promedios.get)
    materia_mas_dificil = {nombre_dificil: promedios[nombre_dificil]}
    return [materia_mas_facil,materia_mas_dificil]

def mostrar_reporte(tabla_global):
    alumnos = tabla_global["boletas_alumnos"]
    estadisticas = tabla_global["estadisticas_materias"]
    facil = tabla_global["materia_mas_facil"]
    dificil = tabla_global["materia_mas_dificil"]

    materias = ["Matematicas", "Fisica", "Quimica", "Historia", "Promedio"]
    n_materias = len(materias)
    n_alumnos = len(alumnos)
    datos = []
    nombres = []
    for alumno in alumnos:
        cal = [
            alumno["matematicas"],
            alumno["fisica"],
            alumno["quimica"],
            alumno["historia"],
            float(alumno["promedio"])
        ]
        datos.append(cal)
        nombres.append(alumno["nombre"])
    x = np.arange(n_alumnos)
    ancho = 0.15
    fig, ax = plt.subplots(figsize=(12, 6))
    for i, materia in enumerate(materias):
        y = [datos[j][i] for j in range(n_alumnos)]
        ax.bar(x + i * ancho, y, width=ancho, label=materia)
        for j, val in enumerate(y):
            ax.text(x[j] + i * ancho, val + 0.5, str(round(val, 1)), ha='center', fontsize=8)

    ax.set_xticks(x + ancho * 2)  # centrar etiquetas
    ax.set_xticklabels(nombres, rotation=45, ha='right')
    ax.set_ylabel("CalificaciÃ³n")
    ax.set_title("Reporte Global de Calificaciones")
    ax.set_ylim(0, 100)
    ax.legend()
    facil_nombre = list(facil.keys())[0]
    facil_prom = list(facil.values())[0]
    dificil_nombre = list(dificil.keys())[0]
    dificil_prom = list(dificil.values())[0]

    ax.text(0, 95, f"Materia mÃ¡s fÃ¡cil: {facil_nombre} ({facil_prom:.2f})", fontsize=10, color='green')
    ax.text(0, 90, f"Materia mÃ¡s difÃ­cil: {dificil_nombre} ({dificil_prom:.2f})", fontsize=10, color='red')

    plt.tight_layout()
    plt.show()


RUTA = r"calificaciones_competo.csv"
ENCABEZADOS = ["nombre","matematicas","fisica","quimica","historia","promedio"]
BOLETA = abrir_generar_archivo(RUTA)

for alumno in BOLETA:
    alumno["matematicas"] = int(alumno["matematicas"])
    alumno["fisica"] = int(alumno["fisica"])
    alumno["quimica"] = int(alumno["quimica"])
    alumno["historia"] = int(alumno["historia"])
    if "promedio" not in alumno:
        alumno["promedio"] = promedio(alumno)
    else:
        pass


while True:
    try:
        usuario = int(input(f"BASE DE DATOS DE ESTUDIANTES\n"
                            "Â¿Que acciÃ³n quiere realizar?\n"
                            "1) Cargar lista de calificaciones globales\n"
                            "2) Cargas boleta individual \n"
                            "3) Actualiza base de datos\n "))
    except ValueError:
        print("Valor equivocado, vuelva a intentar")

    if usuario == 1:
        resultado = boleta_global(BOLETA=BOLETA)
        mostrar_reporte(resultado)
    elif usuario == 2:
        boleta_individual(BOLETA)
    elif usuario == 3:
        actualizar_base_de_datos(BOLETA=BOLETA,ruta=RUTA,campos=ENCABEZADOS)
```

3. Identifique dÃ­as con temperaturas extremas (Â± 2 desviaciones estÃ¡ndar)
4. Muestre un grÃ¡fico de barras de las temperaturas

```python
import matplotlib.pyplot as plt

temperaturas = {"lunes":0,"martes":0,"miercoles":0,"jueves":0,"viernes":0,
                "sabado":0,"domingo":0}
datos = True
while datos:
    try:
        print("Ingresa temperaturas para su analisis")
        temperaturas["lunes"] = int(input("Ingrese la temperatura del Lunes: "))
        temperaturas["martes"] = int(input("Ingrese la temperatura del Martes: "))
        temperaturas["miercoles"] = int(input("Ingrese la temperatura del Miercoles: "))
        temperaturas["jueves"] = int(input("Ingrese la temperatura del Jueves: "))
        temperaturas["viernes"] = int(input("Ingrese la temperatura del Viernes: "))
        temperaturas["sabado"] = int(input("Ingrese la temperatura del Sabado: "))
        temperaturas["domingo"] = int(input("Ingrese la temperatura del Domingo: "))
        datos = False
    except ValueError:
        print("Error. Intentalo nuevamente")

media = sum(temperaturas.values())/len(temperaturas.values())
print(media)
temperaturaMax= max(temperaturas.values())
temperaturaMin= min(temperaturas.values())
temperaturas2 = sorted(temperaturas.values())
print(temperaturas2)
varianza = sum((x - media) ** 2 for x in temperaturas2) / len(temperaturas2)
desviacion_estandar = varianza ** 0.5
limite_superior = media + 2 * desviacion_estandar
limite_inferior = media - 2 * desviacion_estandar

n = len(temperaturas)

if n % 2 == 0:
    mediana = (temperaturas2[n//2 - 1] + temperaturas2[n//2]) / 2
else:
    mediana = temperaturas2[n//2]

dias_extremos = {dia:temp for dia,temp in temperaturas.items() if temp > limite_superior or temp < limite_inferior}
print(dias_extremos)

dias = list(temperaturas.keys())
temps = list(temperaturas.values())
colores = ["red" if dia in dias_extremos.keys() else "blue" for dia in dias]

print("Media:", media)
print("Mediana:", mediana)
print("MÃ¡xima:", temperaturaMax)
print("MÃ­nima:", temperaturaMin)
print("DÃ­as extremos:", dias_extremos)

plt.bar(dias,temps,color=colores)
plt.title("Temperaturas de la semana")
for i, t in enumerate(temps):
    plt.text(i, t, str(t), ha="center", va="bottom")
plt.show()
```
